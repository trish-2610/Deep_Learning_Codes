## **House Price Prediction**

```python
## import required libraries
import numpy as np
from keras.models import Sequential
from keras.layers import Dense,Input

## Training data
## X = house size
## y = house price
X = np.array([500,800,1000,1500,2000])
y = np.array([150000,200000,250000,300000,400000])

## Normalaize data (better training)
X = X / 2000
y = y / 400000

## create model
model = Sequential()

## Input layer
model.add(Input(shape=(1,)))

## Hidden layer
model.add(Dense(10,activation="relu"))

## Output layer
model.add(Dense(1))

## compile model
model.compile(optimizer="adam",loss="mean_squared_error")

## Train model
model.fit(X,y,epochs=300,verbose=1)

## Prediction
test = np.array([[1200]])/2000
prediction = model.predict(test)

## Converting back to actual price
predicted_price = prediction * 400000

print("Predicted Price = ",predicted_price)
```

## **Sentiment Analysis System (LSTM)**

```python
## import required libraries 
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Input , LSTM, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

## Training Data 
## X = product reviews 
## y = sentiment (1 = positive , 0 = negative)

X = ["good quality","bad quality", "excellent item", "not worth", "very good", "poor design"]
y = np.array([1,0,1,0,1,0])

## Tokenization (Text -> numbers)
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
X_seq = tokenizer.texts_to_sequences(X)

## Padding (Same length)
X_pad = pad_sequences(X_seq,maxlen=3)

## creating model
model = Sequential()

## Embedding layer (learn word representations)
model.add(Embedding(input_dim=50,output_dim=8))

## LSTM layer 
model.add(LSTM(10))

## Output layer 
model.add(Dense(1,activation="sigmoid"))

## compile model 
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])

## train model 
model.fit(X_pad,y,epochs=200,verbose=1)

## Prediction 
test = ["very good quality product"]
test_seq = tokenizer.texts_to_sequences(test)
test_pad = pad_sequences(test_seq,maxlen=3)

prediction = model.predict(test_pad)

print("Sentiment score = ",prediction)

if prediction > 0.5:
    print("Positive Review")
else:
    print("Negative Review")
```

## **Simple RNN**

```python
## import required libraries 
import numpy as np 
from keras.models import Sequential
from keras.layers import Dense , Input , SimpleRNN

## Training data 
X = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5]
])
y = np.array([4,5,6])

## Reshape 
# (samples, timesteps, features) = (3, 3, 1)
# Now model understands:
# 3 samples
# each sample has 3 time steps
# each step has 1 feature
X = X.reshape((3,3,1))

## creating model 
model = Sequential()

## Input layer 
model.add(Input(shape=(3,1)))

## RNN layer 
model.add(SimpleRNN(10))

## Output layer 
model.add(Dense(1))

## compile model 
model.compile(optimizer="adam",loss="mean_squared_error")

## train Model 
model.fit(X,y,epochs=300,verbose=0)

## Prediction 
test = np.array([[4,5,6]]).reshape((1,3,1))
prediction = model.predict(test)

print("Next predicted word = ",prediction)
```

## **CNN**

```python
## import required libraries 
import numpy as np 
from keras.models import Sequential
from keras.layers import Dense , Conv2D , MaxPooling2D , Flatten
from tensorflow.keras.datasets import mnist

## load dataset 
(X_train,y_train) , (X_test,y_test) = mnist.load_data()
```

```python
## reshape data (samples, height, width, channels)
X_train = X_train.reshape((-1,28,28,1))
X_test = X_test.reshape((-1,28,28,1))

## Normalize 
X_train = X_train / 255
X_test = X_test / 255

## create model 
model = Sequential()

## CNN layers 
model.add(Conv2D(32,(3,3),activation="relu"))
model.add(MaxPooling2D((2,2)))

## flatten 
model.add(Flatten())

## Dense layers 
model.add(Dense(128,activation="relu"))
model.add(Dense(10,activation="softmax"))

## compile 
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])

## train 
model.fit(X_train,y_train,epochs=5)

## evaluate 
model.evaluate(X_test,y_test)
```
